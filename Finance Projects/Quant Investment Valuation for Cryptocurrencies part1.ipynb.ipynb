```python
# Função para sinalizar quando comprar e quando vender o ativo
def buy_sell(signal):
    buy = []
    sell = []
    flag =  - 1

    for i in range(0,len(signal)):
        if signal['MACD'][i] > signal['Signal Line'][i]:
            sell.append(np.nan)

            if flag != 1 :
                buy.append(signal['Preço'][i])
                flag = 1
            else:
                buy.append(np.nan)

        elif signal['MACD'][i] < signal['Signal Line'][i]:
            buy.append(np.nan)

            if flag != 0 :
                sell.append(signal['Preço'][i])
                flag = 0
            else:
                sell.append(np.nan)
        else:
            buy.append(np.nan)
            sell.append(np.nan)

    return(buy,sell)
```

## 5.2 Análise do Índice de Sharpe <a id="5-2"></a>

### Desde a criação do índice de Sharpe por William Sharpe em 1966, tem sido uma das medidas de risco/retorno mais referenciadas em finanças, e grande parte de sua popularidade é atribuída à sua simplicidade.

### O índice descreve quanto retorno (rx) em excesso você recebe pela volatilidade extra que você suporta por manter um ativo mais arriscado. Lembre-se, você precisa de compensação pelo risco adicional que você assume por não manter um ativo livre de risco.

### A fraqueza subjacente do índice é que nem todos os retornos dos ativos são normalmente distribuídos.

### A taxa de retorno livre de risco (rf) é usada para ver se você é devidamente compensado pelo risco adicional assumido com o ativo. Tradicionalmente, a taxa de retorno livre de risco é a letra do Tesouro dos EUA com a data mais curta.

### Agora que calculamos o excesso de retorno subtraindo a taxa de retorno livre de risco do retorno do ativo de risco, precisamos dividi-lo pelo desvio padrão StdDev (x) do ativo de risco medido. Como mencionado acima, quanto maior o número, melhor será o investimento de uma perspectiva de risco/retorno.

### 5.4 Simulação de Monte Carlo - VaR e CVar para o projeto Cardano<a id='5-4'></a>

```python
df1 = ada_usd_2019_2021
df1 = df1.drop(df1.columns[[0,1,2,3,5,6,7,8]], axis=1)
df1.columns = ['ADA-USD']
df1.head(3)
```

```python
returns = df1.pct_change()
meanReturns = returns.mean()
covMatrix = returns.cov()
```

```python
weights = np.random.random(len(meanReturns))
weights /= np.sum(weights)
```

```python
# Monte Carlo Method
mc_sims = 400 # number of simulations
T = 730 #timeframe in days

meanM = np.full(shape=(T, len(weights)), fill_value=meanReturns)
meanM = meanM.T

portfolio_sims = np.full(shape=(T, mc_sims), fill_value=0.0)

valor_inicial_em_USD = 10000

for m in range(0, mc_sims):
    # MC loops
    Z = np.random.normal(size=(T, len(weights)))
    L = np.linalg.cholesky(covMatrix)
    dailyReturns = meanM + np.inner(L, Z)
    portfolio_sims[:,m] = np.cumprod(np.inner(weights, dailyReturns.T)+1)*valor_inicial_em_USD
```

```python
plt.figure(figsize=(15,8))
plt.plot(portfolio_sims)
plt.ylabel('Valores em Milhões de Dólares')
plt.xlabel('Dias')
plt.title('Simulação de Monte Carlo para o Projeto Cardano')
plt.show()
```

```python
def mcVaR(returns, alpha=5):
    """ Input: pandas series of returns
        Output: percentile on return distribution to a given confidence level alpha
    """
    if isinstance(returns, pd.Series):
        return np.percentile(returns, alpha)
    else:
        raise TypeError("Expected a pandas data series.")
```

```python
def mcCVaR(returns, alpha=5):
    """ Input: pandas series of returns
        Output: CVaR or Expected Shortfall to a given confidence level alpha
    """
    if isinstance(returns, pd.Series):
        belowVaR = returns <= mcVaR(returns, alpha=alpha)
        return returns[belowVaR].mean()
    else:
        raise TypeError("Expected a pandas data series.")
```

```python
portResults = pd.Series(portfolio_sims[-1,:])

VaR = valor_inicial_em_USD - mcVaR(portResults, alpha=5)
CVaR = valor_inicial_em_USD - mcCVaR(portResults, alpha=5)

print('VaR ${}'.format(round(VaR,2)))
print('CVaR ${}'.format(round(CVaR,2)))
```

```python
print('A partir da simulação de Monte Carlo pode-se constatar que o projeto Cardano obteve valores ótimos para Value at Risk e Conditional Value at Risk em comparação aos possíveis ganhos no longo prazo.')
print('Com um VaR de: %.2f' %(VaR))
print('E um CVaR de: %.2f' %(CVaR))
```

### 6.1 Avaliação por Múltiplos <a id="6-1"></a>

### A avaliação por múltiplos pode ser considerada uma metodologia de avaliação baseada no conceito de valor relativo.
### Isso significa dizer que, ao precificar uma empresa, são utilizados múltiplos de outras ações similares com o intuito de estimar qual deveria ser o valor da empresa analisada.
### Ou seja, o valor de uma empresa é baseado na forma como seus pares são precificados no mercado.
### Partimos do pressuposto de que o mercado, em média, precifica corretamente suas cryptomoedas e tais índices podem nos oferecer informações importantes sobre como deveriam estar precificadas no mercado.
### Uma vez que temos a média de mercado para os índices, podemos estimar qual deveria ser o valor das cryptomoedas para que seus múltiplos estejam de acordo com a média do mercado.

### 6.2 CAPM - Capital Asset Pricing Model <a id="6-2"></a>


O Capital Asset Pricing Model (CAPM), ou Modelo de Precificação de Ativos Financeiros, é um método que procura analisar a relação entre o risco e o retorno esperado de um investimento.

Se baseia na soma de dois fatores, a taxa livre de risco e o retorno adicional exigido para compensar o risco assumido, chamado de prêmio de risco. 


<img src='Images/Fórmula CAPM.jpg'>


Para investidores em ações, o valor beta de uma carteira é uma medida de risco muito importante.

Um valor acima de 1 significa que o investimento é mais volátil do que o mercado, um valor inferior a 1 é menos volátil.

Como a maioria das altcoins ainda é vendida e comprada usando Bitcoin (dominância de 44.1%) e Ethereum (dominância 18.2%) as cryptomoedas passam a possuir uma volatilidade muito parecida.

Logo, considerei um beta igual 1.
